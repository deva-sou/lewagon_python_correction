## NumPy Arrays

Arrays are very similar to lists, but they are a couple of differences:

- Arrays are much more computationally efficient
- Arrays can only store one type of data
- Arrays can be multi dimensional

```python
# simple array
arr1 = np.array([1, 2, 3, 4, 5])
print(arr1)

# creating an array from a range
arr2 = np.arange(0, 10, 2)
print(arr2)
```

Once we have some arrays, we can try some operations! It's really simple, no for loops needed!

```python
sum = arr1 + arr2
diff = arr1 - arr2

# the arrays must be of same length.

print("sum:", sum)
print(f"difference: {diff}")
```

We can also explore the many numpy math functions

```python
sqrt = np.sqrt(sum)
print("Square root:", sqrt)

log = np.log(sum)
print("Log:", log)
```

### Multi Dimensional Arrays

Unlike lists, arrays can be **multi-dimensional**!

For example, we can create 2D arrays, also known as **matrices**.

```python
matrice = np.array([[1, 2, 3, 10], [4, 5, 6, 11], [7, 8, 9, 12]])
print(matrice)
```

To check the dimension of your array, they are useful methods :

```python
print('Shape of the first array', arr1.shape)
print('Shape of the matrice array', matrice.shape)
```

As you can see, there is one value in the shape of the first array, so its a 1D-array.
There are two values in the shape of the matrice array, therefore it is a 2D-array.

### Masking

When we want to filter the rows of a DataFrame, we use a mask also called boolean masking.

```python
# first we create a condition  according the filtering we wan't to make.
# here, we would like to only keep the store_id 23.
mask_store_id_23 = (df["store_id"] == 23)
```

```python
mask_store_id_23 # this is called a boolean mask.
```

Then we apply this mask to keep only the lines equal to True

```python
# we use the brackets to select the lines according to the mask.
df_stores_id_23 = df[mask_store_id_23]
df_stores_id_23
```

Multiple conditions can be combined in a mask

```python
mask_store_id_23_item_id_56 = (df["store_id"] == 23)
mask_store_id_23_item_id_56
```

```python
mask_store_id_23_item_id_56 = (df["store_id"] == 23) & (df["item_id"] == 56)
df[mask_store_id_23_item_id_56]
```

### Groupby

You can use GROUP BY just like you did in SQL.
Use Groupby to resume info per dimension. In the following example, we calculate the total quantity per store.

1.   You mention first the name of your dataframe
2.   Then you add .groupby and you specify the name of the column you want to group on.
3.   Next, you put in brackets the columns you want to keep
4.   Lastly, you add the aggregation function you want to use.

```python
df
```

```python
a = df.groupby(["store_id"])
a
```

```python
df.groupby(["store_id"])["quantity"].sum()
```

```python
df
```

Get the average for quantity and price at the same time.

Note, we don't need to use `.agg()` here because it's the same function we want apply.

```python
display(df.groupby(["store_id"])[["quantity", "price"]].mean())
display(df.groupby(["store_id"])[["quantity", "price"]].mean())
```

You may also want to sort the output, you will need to use the `sort_values` function



```python
df.groupby(["store_id"])[["quantity", "price"]].mean().sort_values(by=["quantity"], ascending=[False])
```

```python
df_grouped_by = df.groupby(["store_id"])[["quantity", "price"]].mean()
df_gby_sorted_by = df_grouped_by.sort_values(by=["quantity"], ascending=[False])
df_gby_sorted_by
```

## Joins

### Concatenate

```python
df1 = pd.DataFrame({"id": [100, 101, 102],
                    "first_name": ["paul", "victor", "lucie"],
                    "last_name": ["oiseau", "blanc", "misa"]})
df1
```

```python
df2 = pd.DataFrame({"id": [103, 104, 105],
                    "first_name": ["cécile", "ludivine", "kevin"],
                    "last_name": ["hevert", "masseau", "leroi"]})
df2
```

To concatenate two DataFrames, you need to use the `concat` function of pandas, **beware** of the axis parameter, which determines how to concatenate (in rows or columns)

```python
df_concat = pd.concat([df1, df2], axis=0)
df_concat
```

Reordering the names of the indices

```python
df_concat = df_concat.reset_index(drop = False)
```

```python
df_concat = df_concat.rename(columns={"index": "old_index"})
```

```python
df_concat
```

```python
pd.concat([df1, df2], axis=1) # With opposite axis
```

### join

```python
df3 = pd.DataFrame({"id": [103, 100, 101, 112],
                    "age": [28, 42, 87, 55]})
df3
```

**Inner join**, keep only the ids common to both DataFrame

```python
pd.merge(df_concat, df3, on=["id"], how="inner") # you can have multiple columns
# <=> df_concat.merge(df3, on=["id"], how="inner")
```

```python
df_concat.merge(df3, on=["id"], how="inner")
```

**Left join**, keep only the ids of the left DataFrame and we mark the correspondence if there is one!

```python
pd.merge(df_concat, df3, on=["id"], how="left")
```

**Outer join**, keep the union of the ids of the left and right DataFrame with correspondence when it is possible

```python
pd.merge(df_concat, df3, on=["id"], how="outer")
```

## Handling dates in Pandas

Let's look at the type of the date column of the sales data

```python
print(df.loc[0, "date"])
type(df.loc[0, "date"])
```

You can convert a column to date format using functions already implemented in Python.

```python
df["date"] = pd.to_datetime(df["date"], format='%Y%m%d') # format depends on how is written your date!
```

```python
df
```

List of the columns and their corresponding types. Date column is now datetime64 format.

```python
df.dtypes
```

To access to dates attributes, you need to use `.dt` on your date column:

```python
df["date"].dt.day #.month .year
```

It is now very easy to create another column containing the year week of the dates

```python
df["date-year-week"] = df["date"].dt.strftime("%Y-%U")
df
```

You have plenty functions accessible with `.dt`, you can find them [on this webpage](https://pandas.pydata.org/docs/reference/api/pandas.DatetimeIndex.html).
